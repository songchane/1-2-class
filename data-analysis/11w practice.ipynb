```python
import csv                        

f = open('seoul_0914.csv')       
data = csv.reader(f)

h = next(data)               

max_temp = -999
max_date = ''


for row in data:
    if row[4] == '':
        row[4] = -999
        
    row[4] = float(row[4])        
    
    if max_temp < row[4]:
        max_temp = row[4]
        max_date = row[0]
    
f.close()    

print('최고 기온:', max_temp)
print('최고 기온 날짜:' , max_date)
```

```python
import csv

f =open('seoul_0914.csv')
data = csv.reader(f)

h = next(data)
result = []

for row in data:
    if row[4] != '':
        if (row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '25'):
            result.append(float(row[4]))

f.close()
import matplotlib.pyplot as plt
plt.figure(figsize=(10,2))
plt.plot(result,'r')
plt.title('christmas temperature')
plt.xlabel('Day')
plt.ylabel('The highest temperature')
plt.grid(True)
plt.show()
```
