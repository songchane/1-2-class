```python
import matplotlib.pyplot as plt
import csv

f = open('gender_0922.csv')
data = csv.reader(f)

h = next(data)
male = []
female = []

for row in data:
    if '신도림' in row[0]:
        for i in row[3:104]:
            male.append(-int(i))
        for i in row[106:207]:
            female.append(int(i))
            
f.close()

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('신도림동 지역의 남녀 성별 인구 분포')
plt.barh(range(len(male)), male, color='skyblue', label='male')         
plt.barh(range(len(female)), female, color='pink', label='female')    
plt.xlabel('Population')
plt.ylabel('Age')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('gender_0922.csv')
data = csv.reader(f)

h = next(data)
male = []
female = []

name = input('찾고 싶은 지역의 이름을 입력하세요: ')

#쉼표는 int 불가 replace로 제거해야 함

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            i = i.replace(',', '')
            male.append(int(i))
        for i in row[106:207]:
            i = i.replace(',', '')
            female.append(int(i))
            
            
f.close()

#startangle = 시작 각도 지정
#sum = 합하다(파이썬 내장함수)
print('남자 인구수: ' ,sum(male))
print('여자 인구수: ' ,sum(female))

label = ['male', 'female']
color = ['steelblue', 'thistle']

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + ' 지역의 남녀 성별 인구 분포')
plt.pie([sum(male), sum(female)], autopct='%.1f%%' ,labels=label, colors=color, startangle=90 )
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('gender_0922.csv')
data = csv.reader(f)

h = next(data)
male = []
female = []
diff = []

name = input('찾고 싶은 지역의 이름을 입력하세요: ')

#쉼표는 int 불가 replace로 제거해야 함

for row in data:
    if name in row[0:]:
        for idx in range(3, 104):
            diff.append(int(row[idx].replace(',','')) - int(row[idx + 103].replace(',','')))
            
f.close()


plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.bar(range(len(diff)), diff)
plt.title(name + ' 지역의 남녀 성별 인구 분포')
plt.xlabel('Age')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
#산점도 = point = scatter plot
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []

for num in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10,100))


# s = size, c=color, cmp=color map, alpha = 투명도 설정 (0~1)
plt.scatter(x, y, s=size, c=size, cmap='terrain', alpha=0.7)
plt.grid(True)
plt.colorbar()
plt.show()
```
