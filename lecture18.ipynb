```python
import csv
import matplotlib.pyplot as plt
f = open('age.csv')
data = csv.reader(f)

h = next(data)

result = []

for row in data:
    if '신도림' in row[0]:
        for n in row[3:]:
            result.append(int(n))

f.close()

plt.plot(result)
plt.xlabel('Age')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f = open('age.csv')
data = csv.reader(f)

h = next(data)

name = input('인구 구조가 알고싶은 지역을 입력하세요: ')

result = []

for row in data:
    if name in row[0]:
        for n in row[3:]:
            result.append(int(n))

f.close()

plt.rc('font', family='Malgun Gothic')
plt.plot(result)
plt.title(name + '지역의 연령에 따른 인구수')
plt.xlabel('나이')
plt.ylabel('인구수')
plt.grid(True)
plt.show()
```
