```python
import csv
import matplotlib.pyplot as plt

f = open('seoul_0914.csv')
data = csv.reader(f)

header = next(data)

b = []
c = []

for row in data:
    month = row[0].split('-')[1]
    
    if row[4] != '':
        if month == '11':
            b.append(float(row[4]))
        if month == '08':
            c.append(float(row[4]))
    
f.close()

plt.hist(b, bins=100, color='blue')
plt.hist(c, bins=100, color='red')
plt.title('Temperature Nov and Aug')
plt.grid(True)
plt.show()

```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul_0914.csv')
data = csv.reader(f)

header = next(data)

b = []
jan = []
aug = []

for row in data: 
    
    month = row[0].split('-')[1]
    
    if row[4] != '':
        if month == '01':
            jan.append(float(row[4]))
        elif month == '08':
            aug.append(float(row[4]))

    
f.close()

plt.boxplot([jan, aug])
plt.grid(True)
plt.show()

```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul_0914.csv')
data = csv.reader(f)

header = next(data)

result =[ [], [], [], [], [], [], [], [], [], [], [], [] ]

for row in data: 
    if row[4] != '':
    
        index = int(row[0].split('-')[1]) - 1
        result[index].append(float(row[4]))
    
    
f.close()

plt.boxplot(result)
plt.grid(True)
plt.show()

```
